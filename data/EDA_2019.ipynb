```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

year=2019
df=pd.read_csv('data_clean/scoretable-{}-clean.csv'.format(year))
probs=df.loc[:,'p1':'p5']
sns.swarmplot(data=probs)
```

```python
increasing=(probs.p1<=probs.p2)&(probs.p2<=probs.p3)&(probs.p3<=probs.p4)&(probs.p4<=probs.p5)
decreasing=(probs.p1>=probs.p2)&(probs.p2>=probs.p3)&(probs.p3>=probs.p4)&(probs.p4>=probs.p5)
display(df[increasing])
display(df[decreasing])
print(df[decreasing].shape)
```

```python
df['sex']=df.sex.str.replace(' ','')
```

```python
df.groupby('sex').mean()
```

```python
male=df[df.sex=='M'].total
female=df[df.sex=='F'].total
diff_means=male.mean()-female.mean()
print(len(male),len(female), diff_means)
```

```python
reps=np.empty(50000)
for j in range(50000):
    perm=np.random.permutation(df.total)
    r_male=perm[:len(male)]
    r_female=perm[len(male):]
    reps[j]=r_male.mean()-r_female.mean()
    
print(reps.max(),reps.min())
plt.hist(reps,bins=20)
plt.show()
```

```python
df['mean']=np.mean(probs, axis=1)
df['std']=np.std(probs, axis=1)
```

```python
df[(df['mean']>4)&(df['mean']<4.5)&(df['std']<2.6)]
```

```python
plt.scatter(df['mean'],df['std'])
plt.xlabel('Mean')
plt.ylabel('Standard Deviation')
plt.title('Mean vs. STD')
```
